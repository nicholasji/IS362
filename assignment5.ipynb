What is the northernmost airport in the United States?

```python
#importing pandas and numpy
import pandas as pd
import numpy as np
#retrieving csv from github user hadley
airports = pd.read_csv('https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/airports.csv')
#processing data to get answer, displaying 5 

airports.sort_values(by = 'lat', ascending = False).head(5)
```

What is the easternmost airport in the United States?

```python
#retrieving csv from github user hadley
ea = pd.read_csv('https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/airports.csv')

northeast_airports = ea[(ea['lat'] > 44) & (ea['lat'] <= 50) & (ea['lon'] >= -75) & (ea['lon'] <= -65)]
northeast_airports.sort_values('lon', ascending=False).head(5)

```

The easternmost airport in the United States is Eastport Municipal Airport.
This can be confirmed here:

On February 12th, 2013, which New York area airport had the windiest weather?

```python
wind = pd.read_csv("https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/weather.csv")
wind[(wind["month"]==2) & (wind["day"]==12) & (wind["year"]==2013)].sort_values(by=["wind_speed"], ascending=False).head(5)
```
