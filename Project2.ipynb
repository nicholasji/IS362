#### Cleaning data

```python
import pandas as pd
import numpy as np
import re

ccities = pd.read_csv(r'C:\Users\Nick\Desktop\california_cities.csv')
ccities.head(10)
```

The Unamed coloumn appears to be uncessary. I will drop the coloumn.

```python
ccities_cols =['unnamed', 'city', 'latd', 'longd', 'elevation_m', 'elevation_ft',
       'population_total', 'area_total_sq_mi', 'area_land_sq_mi',
       'area_water_sq_mi', 'area_total_km2', 'area_land_km2', 'area_water_km2',
       'area_water_percent']
ccities.columns = ccities_cols
ccities.head()
ccities.drop('unnamed', axis = 1, inplace = True)
ccities.head()
```

The cities listed under city do not contain spaces where necessary. For example AgouraHills should be Agoura Hills. 
I will add the space.

```python
def spacing(city):
    space = re.findall('[A-Z][^A-Z]*', city)
    return ' '.join(space)

ccities['city'] = ccities['city'].apply(spacing)
ccities
```

#### Which cities have the highest population total?


```python
ccities[['city', 'population_total']].sort_values('population_total', ascending=False).head(5)
```

#### Which cities have the lowest area of water?

```python
ccities[['city', 'area_water_percent']].sort_values('area_water_percent', ascending=True).head(5)
```

#### Which cities have the highest area of water?

```python
ccities[['city', 'area_water_percent']].sort_values('area_water_percent', ascending=False).head(5)
```

### NYC Death Data

```python
deaths = pd.read_csv(r'C:\Users\Nick\Desktop\Death_Rates_NYC.csv')
deaths.head(10)
```

Cleaning data, replacing . with null values, dropping uncessary data and reassigning data from string to int

```python
deaths = deaths.replace(to_replace='.', value=np.nan)
deaths.isnull().sum()
deaths.Deaths = deaths.Deaths.dropna().astype(np.int64)
deaths.sort_values(['Year', 'Deaths'], ascending=[True,False]).head(10)
```

Showing deaths for White Non-Hispanic Females.

```python
deaths[(deaths['Sex']=='F') & (deaths['Race Ethnicity'] == 'White Non-Hispanic')].sort_values(by = ['Year' , 'Deaths'])
```
