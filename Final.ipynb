5. Valence - The higher the value, the more positive mood for the song.

7. Acoustic - The higher the value the more acoustic the song is.

```python
sb.lmplot(y ='danceability',x = 'instrumentalness',data = sdatanew)
```

** Relationship between danceability and instrumentalness:** The higher the instrumentalness value for a track, the greater likelihood that it contains no vocal content. As we can see from the chart above, a song with low vocal content will more than likely have a higher in danceability. 


```python
sb.lmplot(y ='valence',x = 'energy',data = sdatanew)
```

** Relationship between valence and energy: **Surprisingly, the sweet spot for happy songs are not necessiarly high in energy. It is between 0.6 and 0.8 energy.

```python
sb.lmplot(y ='valence',x = 'tempo',data = sdatanew)
```

**Relationship betweenvalence and tempo:** The majority of the songs on this list lie between 90 to 140 BPM. This range also expresses the most variation in emotion. It's interesting to note that the saddest three songs are 120BPM followed by 180BPM and 200BPM.
