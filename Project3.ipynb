Create a pandas DataFrame that displays Customers’ Last Name and First Name, and each customer’s purchased Track names and 
Album Titles. The information should be sorted by Customer LastName then Customer FirstName.  

```python
import  pandas as pd
import numpy as np
import sqlite3


sql = '''SELECT Customer.LastName, Customer.FirstName, Track.Name, Album.Title
         FROM Customer
         INNER JOIN Invoice 
         ON Customer.CustomerId = Invoice.CustomerId
         INNER JOIN InvoiceLine
         ON Invoice.InvoiceId = InvoiceLine.InvoiceId
         INNER JOIN Track
         ON InvoiceLine.TrackId = Track.TrackId
         INNER JOIN Album
         ON Track.AlbumId = Album.AlbumId
         ORDER BY Customer.LastName, Customer.FirstName'''

con = sqlite3.connect(r"C:\Users\Nick\Desktop\ChinookDatabase1.4_CompleteVersion\Chinook_Sqlite.sqlite")
db = pd.read_sql_query(sql, con)
pd.read_sql_query(sql , con=con).head(39)
```
